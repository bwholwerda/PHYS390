```python
# load up the various python libraries we need. 
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from astropy.io import ascii
from astropy.io import fits
from astropy.io import ascii

# load data table
data = ascii.read("gasrichdatatable.txt",names=['Galaxy','D','Vc','errVc','logM*','errM*','logMg','errMg','Ref.'])  
print(data)

# Galaxy - name of the galaxy
# D - distance to the galaxy in MPc,
# Vc - rotation velocity (km/s)
# errVc - error in the rotation velocity
# logM* - log of stellar mass
# errM* - error in the stellar mass
# logMg - log of the gas mass
# errMg - error in the gas mass
# Ref. -  reference for these values

Mstar = 10**data['logM*']
```

## 1. Plot any of the three intrinsic qualities of the galaxies (logM\*, logMg or vc) against distance (D).

```python
# student work below 
plt.figure()
ax = plt.subplot(111)
ax.plot(data['D'], data['Vc'], 'k.',label='galaxy')
xlabel('D (Mpc)')
ylabel('Vc (km/s)')
#ax.set_ylim([0.0, 10000.0])
legend()
plt.savefig('D-vc.png',bbox_inches='tight')


```

## 2. Plot the stellar mass (derived from the total stellar luminosity) against the log(velocity)

```python
# student work below 
plt.figure()
plt.plot(Mstar, data['Vc'], 'k.',label='galaxy')
plt.xlabel('logM* ')
plt.ylabel('Vc (km/s)')
#ax.set_ylim([0.0, 10000.0])
plt.legend()
plt.savefig('Mstar-vc.png',bbox_inches='tight')
```

## 3. Plot the T-F relation: log(M*) vs Vc 

```python

# student work below 
plt.figure()
plt.plot(data['logM*'], data['Vc'], 'k.',label='galaxy')
plt.xlabel('logM*')
plt.ylabel('Vc')
legend()
plt.savefig('T-F.png',bbox_inches='tight')


plt.figure()
plt.errorbar(data['logM*'], data['Vc'] , data['errVc'],  data['errM*'], 'k.',label='galaxy')
plt.xlabel('logM*')
plt.ylabel('Vc')
#ax.set_ylim([0.0, 10000.0])
legend()
plt.savefig('T-F_errors.png',bbox_inches='tight')

```

## 4. Is this consistent with a linear relation? Does there appear to be intrinsic scatter in this relation? Over how many orders of magnitude does any scaling relation hold?

*your answer here*

This is a linear relationship but intrinsic scatter is indeed still present. 

## 5. Plot log gas and stellar mass against each other. 

```python
# student work below 

plt.figure()
plt.plot(data['logM*'], data['logMg'], 'k.',label='galaxy')
plt.xlabel('logM*')
plt.ylabel('Vc')
legend()
plt.savefig('stars-gas-mass.png',bbox_inches='tight')

```

## 6. Add gas and stellar mass together into log(M8+Mg) and plot against Vc 

```python
# student work below 
plt.figure()
plt.plot(data['logM*'] , data['Vc'], 'g.',label='galaxy')
plt.plot(np.log10(10**data['logM*'] + 10** data['logMg']), data['Vc'], 'k.',label='galaxy')
plt.xlabel('log(M*+Mg)')
plt.ylabel('log(Vc)')
legend()
plt.savefig('Baryonic_TF.png',bbox_inches='tight')

```
